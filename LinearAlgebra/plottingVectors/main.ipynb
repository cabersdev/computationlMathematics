```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import tikzplotlib as tk
from IPython.display import Math,display
```

```python
# creating a row vector
row_vec = np.array([1,3])

# creating a column vector
col_vec = np.array([ [-1],[-3] ])

# displaying vectors
display(Math(sym.latex(sym.sympify(row_vec))))
display(Math(sym.latex(sym.sympify(col_vec))))
```

```python
# plotting 2D geometric vectors
plt.plot([0,row_vec[0]],[0,row_vec[1]],'r',label='row vector')
#plt.plot([0,col_vec[0]],[0,col_vec[0]],'b',label='column vector')

plt.axis('square')
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()

plt.show()

# saving the plot in a latex document (comment plt.show() and plt.legend() before running tk.save())
#tk.save('twoDimensionalVector.tex')
```

```python
# plotting a 3D geometric vectors

from mpl_toolkits.mplot3d import Axes3D

# declaring the vector
v = np.array([3,0,-4])
w = np.array([-1,1,3])

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

ax.plot([0,v[0]],[0,v[1]],[0,v[2]],'b', linewidth=3,label='v')
ax.plot([0,w[0]],[0,w[1]],[0,w[2]],'r' ,linewidth=3, label='w')

ax.set_xlim3d([-5,5])
ax.set_ylim3d([-5,5])
ax.set_zlim3d([-5,5])

ax.plot([-5,5],[0,0],[0,0],'--',color=[.7,.7,.7])
ax.plot([0,0],[-5,5],[0,0],'--',color=[.7,.7,.7])
ax.plot([0,0],[0,0],[-5,5],'--', color=[.7,.7,.7])
```
