```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Math,display
```

```python
# multiplying scalar

# declaring vector v
v = np.array([1,-3])

# declaring a collection of scalar
s = [-3,4,-1,2]

for i in s:
    sv = i*v
    plt.plot([0,sv[0]],[0,sv[1]],'o-',linewidth=3,label='$\\lambda%g$' %i)

plt.legend() 
plt.axis([-5,5,-5,5])
plt.axis('square')
plt.grid()
plt.show()
```

```python
# adding a vector to a vector

# declaring two vectors
v = np.array([1,3])
w = np.array([4,1])

# adding vectors

# sv = v + w

sv = np.add(v,w)

#sv = np.zeros(2)
#for i in range(0,2):
#    sv[i] = v[i] + w[i]

plt.plot([0,v[0]],[0,v[1]],'r',label='v')
plt.plot([0,w[0]],[0,w[1]],'b',label='w')
plt.plot([0,sv[0]],[0,sv[1]],'g',label='v+w')

plt.axis([-5,5,-5,5])
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
# adding vectors visualize geometrically

plt.plot([0,v[0]],[0,v[1]],'r',label='v')
plt.plot([0,w[0]]+v[0],[0,w[1]]+v[1],'b',label='w')
plt.plot([0,sv[0]],[0,sv[1]],'g',label='v+w')

plt.axis([-5,5,-5,5])
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
# adding vectors visualize geometrically
plt.plot([0,v[0]],[0,v[1]],'r',label='v')
plt.plot([0,w[0]],[0,w[1]],'b',label='w')
plt.plot([0,v[0]]+w[0],[0,v[1]]+w[1],'--')
plt.plot([0,w[0]]+v[0],[0,w[1]]+v[1],'--')
plt.plot([0,sv[0]],[0,sv[1]],'g',label='v+w')

plt.axis([-5,5,-5,5])
plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```
